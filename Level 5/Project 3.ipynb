Titanic Survival Insights

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\loalo\\AppData\\Local\\Temp\\dc91df6f-be6b-46b7-b4a8-9873fd46ca4a_titanic.zip.a4a\\Titanic-Dataset.csv")
df
```

```python
summary_stats = df[["Age","Fare"]].describe()
summary_stats
```

```python
survival_rates = df.groupby("Pclass")["Survived"].mean()
survival_rates
```

```python
sns.barplot(x = survival_rates.index, y = survival_rates.values)
plt.xlabel("Passenger class")
plt.ylabel("Survival Rate")
plt.title("Survival rats Passenger class")
```

```python
plt.show()
```
