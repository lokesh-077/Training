```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\loalo\\Downloads\\housing.csv")
df
```

```python
summary_stats = df[["price","area"]].describe()
summary_stats
```

```python
price_col = "price"
area_col = "area"
```

```python
df[price_col] = pd.to_numeric(df[price_col],errors='coerce')
df[area_col] = pd.to_numeric(df[area_col],errors='coerce')
```

```python
df['price_per_sqft'] = df[price_col] / df[area_col]
```

```python
df.head()
```

```python
plt.scatter(df["area"],df["price"])
plt.xlabel("area")
plt.ylabel("price")
plt.title("House Price vs Area")
plt.show()
```
