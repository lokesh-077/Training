Customer Segmentation Basics

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('shopping_trends_updated.csv',encoding='latin1')
df
```

```python
purchase_amnt = df["Purchase Amount (USD)"]
purchase_amnt
```

```python
purchase_amnt_mean = df["Purchase Amount (USD)"].mean()
purchase_amnt_meadian = df["Purchase Amount (USD)"].median()
purchase_amnt_mode = df["Purchase Amount (USD)"].mode()[0]
print(f"Mean : {purchase_amnt_mean:.2f},Median : {purchase_amnt_meadian:.2f},Mode:{purchase_amnt_mode}")
```

```python
high_segment = df["Purchase Amount (USD)"] > purchase_amnt_mean
low_segment = df["Purchase Amount (USD)"] <=purchase_amnt_mean
print(f"High Spenders: {len(high_segment)} customers")
print(f"Low Spenders: {len(low_segment)} customers")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=["High Spenders", "Low Spenders"], y=[len(high_segment), len(low_segment)], palette="Blues")
plt.ylabel("Number of Customers")
plt.title("Customer Segmentation by Spending")
plt.show()
```
