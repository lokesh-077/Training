```python
df = df[["price", "area"]].dropna() 

print(df.describe())
```

```python
pearson_corr = df["price"].corr(df["area"], method="pearson")

spearman_corr = df["price"].corr(df["area"], method="spearman")

print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["area"], y=df["price"], alpha=0.7)
```

```python
plt.title("Scatter Plot: House Price vs. Size")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($)")
```
