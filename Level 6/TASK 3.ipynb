```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
 
df = pd.read_csv( "C:\\Users\\loalo\\Downloads\\Employee Attrition.csv" )
df
```

```python
print("Dataset Overview:")
print(df.head())
```

```python
df_cleaned = df.dropna(subset=["dept"])
sample_df, _ = train_test_split(df_cleaned, test_size=0.7, stratify=df_cleaned["dept"], random_state=42)
print("\nSampled Data Distribution:\n", sample_df["dept"].value_counts(normalize=True))
print("\nOriginal Data Distribution:\n", df_cleaned["dept"].value_counts(normalize=True))
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
df["dept"].value_counts().plot(kind="bar", color="blue", alpha=0.7)
plt.title("Original Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
```

```python
plt.subplot(1, 2, 2)
sample_df["dept"].value_counts().plot(kind="bar", color="green", alpha=0.7)
plt.title("Sampled Department Distribution (30%)")
plt.xlabel("Dept")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```
