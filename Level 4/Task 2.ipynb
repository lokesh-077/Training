PDF Protection

```python
from pypdf import PdfReader, PdfWriter
```

```python
input_pdf = "C:\\Users\\loalo\\Downloads\\UNIT 5.pdf"
reader = PdfReader(input_pdf)
reader
```

```python
writer = PdfWriter()
writer
```

```python
for page in reader.pages:
    writer.add_page(page)
```

```python
my_pass = "password"
writer.encrypt(my_pass)
```

```python
output_pdf = "output_encrypted.pdf"
with open(output_pdf, "wb") as output_file:
    writer.write(output_file)
```

```python
print("Encryption successful! Encrypted PDF saved as:", output_pdf)
```
