Product Price Analysis

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("shopping_trends_updated.csv")
df
```

```python
price = df["Purchase Amount (USD)"]
price
```

```python
price_range = price.max() - price.min()
price_variance = np.var(price,ddof=1)
price_std_dev = np.std(price,ddof=1)
```

```python
print("Dispersion Statistics for Product Prices:")
print(f"range      : {price_range:.2f}")
print(f"variance   : {price_variance:.2f}")
print(f"std_dev.  : {price_std_dev:.2f}")
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(price, bins=30, kde=True, color="skyblue")
plt.title("Price Distribution - Histogram")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()  
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x=price, color="lightgreen")
plt.title("Price Distribution - Boxplot")
plt.xlabel("Purchase Amount (USD)")
plt.show()
```
